### Question 1

```python
import pandas as pd
Laliga = pd.read_csv('Laliga.csv')
Laliga=Laliga.rename(columns=Laliga.iloc[0]).drop(Laliga.index[0])
Laliga.replace('-', 0, inplace=True)
Laliga.head(10)
```

### Question 2

```python
Laliga["Debut"]= Laliga["Debut"].str.split("-", n = 1, expand = True)
Laliga.Debut.dtypes
Laliga["Debut"]=Laliga["Debut"].apply(pd.to_numeric)
print('The teams which have started playing between 1930-1980')
Laliga.loc[(Laliga['Debut'] >= 1930) & (Laliga['Debut'] <= 1980),['Team','Debut']].sort_values('Debut',ascending = True)
```

### Question 3

```python
a=Laliga.columns
Laliga.dtypes
Laliga[a[0]]=Laliga[a[0]].apply(pd.to_numeric)
Laliga[a[2:17]]=Laliga[a[2:17]].apply(pd.to_numeric)
Laliga[a[19]]=Laliga[a[19]].apply(pd.to_numeric)
Top_5=Laliga.sort_values('Points', axis=0,ascending=False)
print('Top 5 teams in terms of points')
Top_5[['Team','Points']].head(5)
```

### Question 4

```python
def Goal_diff_count(var1,var2):        
    GDC=var1-var2
    return(GDC)

b=Goal_diff_count(Laliga['GoalsFor'],Laliga['GoalsAgainst'])
dd=pd.DataFrame(b)

dd.columns=['Goal_differences']
result = pd.concat([Laliga['Team'],dd], axis=1)
print('The teams with their Goal Differences')
print(result)

Max_goal_diff=result.sort_values('Goal_differences', axis=0,ascending=False)
print('Team with max goal diff is:',Max_goal_diff.iloc[0,0] ,'and Goal difference:',Max_goal_diff.iloc[0,1] )
Min_goal_diff=result.sort_values('Goal_differences', axis=0,ascending=True)
print('Team with min goal diff is:',Min_goal_diff.iloc[0,0] ,'and Goal difference:',Min_goal_diff.iloc[0,1] )
```

### Question 5

```python
Laliga['Winning Percent']=(Laliga['GamesWon']/Laliga['GamesPlayed'])*100
Laliga.replace('NaN', 0, inplace=True)
Top_5_Winning_Percent=Laliga.sort_values('Winning Percent', axis=0,ascending=False)
print('Top 5 Team with maximum winning percent')
Top_5_Winning_Percent[['Team','Winning Percent']].head(5)
```
